# **User Stories**

## **As a ___ I want to ___ so that___**

- Head of E-Commerce
- View revenue trends over time
- can identify seasonality and plan promotional activity

- Business Manager
- compare revenue across countries
- can prioritise high performing markets

- Marketing Manager
- customers grouped by purchasing behavior
- retention strategies can be designed

- Operations Manager
- understand which products are frequently purchased together
- product bundling strategies can be created

- Data Analyst
- data cleaning decisions to be clearly documented
- analysis is reproducible and auditable

- Executive User
- interactive dashboard
- can plan targeted campaigns
